```python
# Displaying the entities of DataFrame for rows where 'Code' is empty or NaN
entities_with_empty_code = df_updateddataset[df_updateddataset['Code'].isnull() | (df_updateddataset['Code'] == '')]['Entity']
distinct_entities = entities_with_empty_code.unique()
print(distinct_entities)
print(len(distinct_entities))
```

```python
#There are total 35 entities whose code is empty or null
```

```python
#Generating the codes for the entities whose code is empty or null
def code_(entity):
    words = entity.split()
    code = ''.join(word[0] for word in words)
    return code.upper()

entities = ['Andean Latin America', 'Australasia', 'Caribbean', 'Central Asia',
            'Central Europe', 'Central Europe, Eastern Europe, and Central Asia',
            'Central Latin America', 'Central Sub-Saharan Africa', 'East Asia',
            'Eastern Europe', 'Eastern Sub-Saharan Africa', 'England', 'High SDI',
            'High-income', 'High-income Asia Pacific', 'High-middle SDI',
            'Latin America and Caribbean', 'Low SDI', 'Low-middle SDI', 'Middle SDI',
            'North Africa and Middle East', 'North America', 'Northern Ireland',
            'Oceania', 'Scotland', 'South Asia', 'Southeast Asia',
            'Southeast Asia, East Asia, and Oceania', 'Southern Latin America',
            'Southern Sub-Saharan Africa', 'Sub-Saharan Africa',
            'Tropical Latin America', 'Wales', 'Western Europe',
            'Western Sub-Saharan Africa']

entity_codes = {entity: code_(entity) for entity in entities}

for entity, code in entity_codes.items():
    print(f"{entity} - {code}")
```

```python
#Replacing the empty rows of the code of corresponding entity with the above generated codes
for entity, code in entity_codes.items():
    df_updateddataset.loc[df_updateddataset['Entity'] == entity, 'Code'] = df_updateddataset.loc[df_updateddataset['Entity'] == entity, 'Code'].fillna(code).replace('', code)

df_updateddataset
```

```python
#Checking column names which contain null values and empty strings rows in the dataframe after updating
empty_nan_columns = df_updateddataset.columns[(df_updateddataset.isnull() | (df_updateddataset == '')).any()].tolist()
print(empty_nan_columns)
```
